**1.d.** Modify the following function so that it can load an RGB image and convert it to a grayscale image (*i.e.* converting it from a `HW3` array to a `HW` array) using the formula `L = 0.2989 * R + 0.5870 * G + 0.1140 * B`, where `R`, `G` and `B` are the three color channels of the color image and `L` is the luminance (*i.e.* the grayscale image). Include a test so that your function also works if the input image is grayscale.

```python
def load_image(name, crop_window=None):
    I = plt.imread(name)
    if crop_window is not None:
        I = I[crop_window[0]:crop_window[1],
              crop_window[2]:crop_window[3]]
    coeff = np.array([0.2989, 0.5870, 0.1140])
    I = I.dot(coeff)
    I = I.astype('float') / 255.
    return I

I = load_image('legumes.jpg')
plt.figure()
plt.imshow(I)
plt.show()
plt.savefig("gray_legumes.jpg")
```
